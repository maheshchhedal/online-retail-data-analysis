```python
#check shpae the data
df.shape
```

```python
#cloumns
df.columns
```

```python
#convert columns name into lowercase
df.columns=df.columns.str.strip().str.lower()
df.columns
```

```python
#drop unusage column
columns_to_drop= [
    'name',
    'email',
    'phone',
    'address',
    'zipcode',
    'product_type',  
    'products'       
]

df = df.drop(columns=columns_to_drop)
```

```python
# check info of the data
df.info()
```

```python
#convert into correct data types
df['date']=pd.to_datetime(df['date'])
df['year']=pd.to_datetime(df['year'])
df['month']=pd.to_datetime(df['month']).dt.month
df['time'] = pd.to_datetime(df['time'], format='%H:%M:%S')
```

```python
#check duplicate
df.duplicated().sum()
#drop duplicate
df=df.drop_duplicates()
df.duplicated().sum()

# Drop duplicate transaction IDs because each transaction should have a unique ID
df = df.drop_duplicates(subset=['transaction_id'])
```

```python
#check nullvalues
df.isnull().sum()
```

```python
missing = df.isnull().sum()

missing_percent = (df.isnull().sum() / len(df)) * 100

print(pd.DataFrame({
    "Missing Values": missing,
    "Percentage": missing_percent
}).sort_values("Percentage", ascending=False))
```

```python
df.dropna(subset=['amount'], inplace=True)
df = df[df['amount'] >= 0]
```

```python
numeric_cols = ['age','ratings', 'total_purchases']

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
# Describe selected numeric columns
df[['age', 'total_purchases', 'amount', 'total_amount', 'ratings']].describe()
```

```python
#save clean data 
df.to_csv('../data/clean/clean.csv',index=False)
```
